```python
df_bus = pd.read_csv('사용년월별_버스정류장_근처지하철_출퇴근_총이동량.csv',index_col=0)
df_bus
```

```python
df_subway = pd.read_csv('df_new.csv',index_col=0, encoding='utf-8')
df_subway
df_출근 = pd.read_csv('../df_2018_출근.csv')
df_퇴근 = pd.read_csv('../df_2018_퇴근.csv')
df = pd.concat([df_출근,df_퇴근])
many_passengers_subway = list(df['지하철역'].drop_duplicates()) # 중복 제거
len(many_passengers_subway)
```

```python
df_subway['총 이동량']=df_subway['출근 시간대 인원']+df_subway['퇴근 시간대 인원']
df_subway
```

```python
df_subway['상위 25% 포함여부'] = 0
for i in tqdm(range(len(df_subway))):
    df_subway['상위 25% 포함여부'][i] = df_subway['지하철역'][i] in many_passengers_subway
    # 한줄씩 추가
df_subway
```

```python
df_subway = df_subway[df_subway['상위 25% 포함여부']==True]
df_subway.describe()
```

```python
df_bus_pivot = df_bus.pivot_table(index=['사용년월','name','lat','lng'],values=['출근','퇴근','총 이동량'],aggfunc=sum).reset_index()
df_bus_pivot
```

```python
df_subway_pivot = df_subway.pivot_table(index=['사용월','지하철역','lat','lng'],values=['출근 시간대 인원','퇴근 시간대 인원','총 이동량'],aggfunc=sum).reset_index()
df_subway_pivot
```

```python
df_subway_pivot.rename(columns = {'사용월' : '사용년월'}, inplace = True)
df_subway_pivot.rename(columns = {'지하철역' :'name'}, inplace = True)
df_subway_pivot.rename(columns = {'출근 시간대 인원' : '출근'}, inplace = True)
df_subway_pivot.rename(columns = {'퇴근 시간대 인원' : '퇴근'}, inplace = True)
df_subway_pivot
```

```python
ds = pd.concat([df_bus_pivot,df_subway_pivot])
```

```python
ds = ds.pivot_table(index=['사용년월','name','lat','lng'],values=['출근','퇴근','총 이동량'],aggfunc=sum).reset_index()
ds
```

```python
ds = ds.astype({'사용년월':'string'})
ds.dtypes
```

```python
df_18 = ds[ds['사용년월'].str.contains('2018')]
df_19 = ds[ds['사용년월'].str.contains('2019')]
df_20 = ds[ds['사용년월'].str.contains('2020')]
df_21 = ds[ds['사용년월'].str.contains('2021')]
```

```python
df_18_19 = pd.concat([df_18,df_19])
df_20_21 = pd.concat([df_20,df_21])
```

```python
df_18_19_pivot = df_18_19.pivot_table(index=['name','lat','lng'], aggfunc=sum).reset_index()
df_18_19_pivot.sort_values(['총 이동량','출근','퇴근'], ascending=False, inplace=True)
df_18_19_pivot
```

```python
df_18_19_pivot.describe()
```

```python
df_20_21_pivot = df_20_21.pivot_table(index=['name','lat','lng'], aggfunc=sum).reset_index()
df_20_21_pivot.sort_values(['총 이동량','출근','퇴근'], ascending=False, inplace=True)
df_20_21_pivot
df_20_21_pivot.describe()
```

# 지도 시각화

```python
total_map = folium.Map(location=[37.53162973623989, 126.98125303943947], zoom_start=12.5)
```

```python
df_20_21_pivot
```

```python
df_18_19_pivot
```

```python
for i in range(len(df_18_19_pivot)):
    folium.Circle(location=(df_18_19_pivot.loc[i]['lat'],df_18_19_pivot.loc[i]['lng']),
                    color='white',
                    radius=2).add_to(total_map)

    num = df_18_19_pivot.loc[i]['총 이동량']/900000

    folium.CircleMarker(location=(df_18_19_pivot.loc[i]['lat'],df_18_19_pivot.loc[i]['lng']),
                        radius=num,
                        fill=True,
                        color='blue',
                        popup=folium.Popup(df_18_19_pivot.loc[i]['name'], max_width=100)).add_to(total_map)
```

```python
for i in range(len(df_20_21_pivot)):
    folium.Circle(location=(df_20_21_pivot.loc[i]['lat'],df_20_21_pivot.loc[i]['lng']),
                    color='white',
                    radius=2).add_to(total_map)

    num = df_20_21_pivot.loc[i]['총 이동량']/900000

    folium.CircleMarker(location=(df_20_21_pivot.loc[i]['lat'],df_20_21_pivot.loc[i]['lng']),
                        radius=num,
                        fill=True,
                        color='red',
                        popup=folium.Popup(df_20_21_pivot.loc[i]['name'], max_width=100)).add_to(total_map)
```
