```python
bus = pd.read_csv('버스_출근퇴근_총_인원정보(최종).csv', index_col=0)
# bus = bus.drop(['Unnamed: 0'], axis=1)
bus
```

```python
bus18 = bus[bus['사용년월'] < 201901]
bus18
```

```python
bus18 = bus18.drop(['노선번호', '노선명', '표준버스정류장ID'], axis=1)
bus18
```

```python
bus18_edit = pd.pivot_table(bus18, index=['사용년월','역명', '위도', '경도'], values=['출근', '퇴근', '총 이동량'], aggfunc='sum')
bus18_edit
```

```python
bus18_edit = bus18_edit.reset_index()
bus18_edit
```

```python
bus18_edit_sum = bus18_edit['총 이동량'].sum()
bus18_edit_sum
```

```python
bus18_month = bus18_edit[['사용년월', '총 이동량', '출근', '퇴근']].groupby(['사용년월'],as_index=False).sum()
bus18_month
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from matplotlib import rc 

plt.rcParams["axes.unicode_minus"] = False 
rc("font", family="Arial Unicode MS") # Windows: Malgun Gothic 
# %matplotlib inline 
get_ipython().run_line_magic("matplotlib", "inline")
```

```python
# 연도별 총 버스 이용객수 라인 그래프
# 2020년에 코로나19가 본격적으로 확산하기 시작하자 그 이전년도 보다 버스 이용이 급격하게 줄어듦 
# 2021년엔 이용객이 다소 많아졌으나 코로나 이전 보다는 못 미치는 수준

x_values = [2018, 2019, 2020, 2021]
y_values = [bus18_edit_sum, bus19_edit_sum, bus20_edit_sum, bus21_edit_sum]
plt.xticks(x_values, x_values)
plt.plot(x_values, y_values, marker='*', linestyle='solid', color='green')
plt.xlabel('연도')
plt.ylabel('버스 이용객수(억명)')
plt.tick_params(axis='x', width=4)
title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('2018년~2021년 서울시 버스 이용객수 추이', fontdict = title_font , pad = 30)
plt.show
```

```python
bus_month = pd.concat([bus18_month, bus19_month, bus20_month, bus21_month], axis=0)
bus_month.to_csv('당장볼것.csv')
```
