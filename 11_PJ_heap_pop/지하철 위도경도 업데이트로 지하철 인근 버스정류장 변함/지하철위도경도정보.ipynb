```python
import pandas as pd
```

```python
df = pd.read_csv("지하철_시간대별_인원정보(위치정보_최종).csv", encoding='cp949', index_col=0)
```

```python
df
```

```python
sub_station_loc = df.pivot_table(index=['호선명','지하철역'],values=['lat','lng']).reset_index()
```

```python
sub_station_loc.rename({'호선명':'line', '지하철역':'name'},axis=1,inplace=True)
sub_station_loc
```

```python
sub_station_loc.to_csv('subway_station_coordinate.csv')
```
