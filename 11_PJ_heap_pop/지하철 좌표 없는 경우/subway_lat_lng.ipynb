```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('지하철_시간대별_인원정보(최종).csv')
```

```python
df.tail()
```

```python
df['lat'].fillna(0, inplace=True)
df['lng'].fillna(0, inplace=True)

df_nan_location = df[df['lat'] == 0]
```

```python
df_update = df_nan_location.drop(columns=["사용월",'07시-08시 하차인원', '08시-09시 하차인원', '09시-10시 하차인원', '17시-18시 승차인원', '18시-19시 승차인원', '19시-20시 승차인원'])
```

```python
df_update.keys()
```

```python
df_update = df_update.pivot_table(index=['지하철역','호선명']).reset_index()
```

```python
df_update['호선명'].unique()
```

```python
df_update.to_csv('지하철_위치정보_없음.csv', encoding='cp949')
```
