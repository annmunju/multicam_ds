# 1. 딥러닝 소개

### 인공지능

- Question "컴퓨터가 생각할 수 있는가?" (1950년대 초기 컴퓨터과학분야)
- 보통의 사람이 수행하는 지능적인 작업을 자동화하기 위한 연구활동
- 심볼릭(symbolic) AI(1950년대~1980년대)
    - **명시적 규칙을 충분히 많이 만들어 지식을 다루면 인간 수준의 인공지능을 만들 수 있다**
    - 전문가시스템(1980년대)
    - 잘 정의된 논리적인 문제를 푸는데 적합
    - 이미지 분류, 음성인식, 언어번역 등의 불분명한 문제을 위한 명확한 규칙을 찾기 어려움
- 새로운 접근: 머신러닝
    - **데이터와 데이터로부터 기대되는 답을 주어 실행 규칙을 찾는 방법**
    - 훈련(training) : 데이터에서 통계적 구조를 찾아 그 작업을 자동화하기 위한 규칙 생성

- 머신러닝은 수리통계와 밀접하게 관련이 되어있지만, 전통적 통계분석 방법에 비해 수학적 이론이 비교적 부족하고 엔지니어링 지향적(실천적 접근, 경험을 바탕으로 증명되는 경우가 많음)

**데이터에서 표현을 학습하기**

- 머신러닝을 위한 필요 요소
    1. 입력 데이터 포인트 : 
        - 음성인식 : 대화가 녹음된 사운드 파일
        - 이미지태깅 : 사진
    2. 기대 출력 : 
        - 음성인식 : 사람이 사운드 파일을 듣고 옮긴 글
        - 이미지 태깅 : 강아지, 고양이 등과 같은 태그
    3. 알고리즘의 성능을 측정하는 방법
        - 알고리즘의 현재 출력과 기대 출력 간의 차이를 결정하기 위해 필요
        - 측정값은 알고리즘의 작동방식을 교정하기 위한 신호로 다시 피드백 됨
            - 이러한 수정단계를 학습 이라고 함

**머신러닝과 딥러닝의 핵심 문제 : 의미있는 데이터로의 변환**   
- 즉, **입력 데이터를 기반으로 기대출력에 가깝게 만드는 유용한 `표현`을 학습**하는 것

## 표현  
- 데이터를 인코딩하거나 묘사하기위해 데이터를 바라보는 다른 방법

**예. 데이터 포인트가 빨간색인지 흰색인지를 출력하는 알고리즘을 개발**
- 입력 : 포인트 좌표(x, y)
- 기대출력 : 포인트 색깔
- 알고리즘 성능 : 정확히 분류한 포인트 비율

- 포인트 중 `정확히 분류된 좌표변환`을 찾는 과정 = **표현**
    - 색깔 분류 문제를 “x > 0인 것은 빨간색 포인트임
    - 또는 “x < 0인 것은 흰색 포인트다.”라는 간단한 규칙
    - 이 분류 문제를 해결한 것은 새로운 표현임

- 머신러닝의 **학습(learning)이란 더 나은 표현으로 바꾸어주는 변환을 찾는 자동화된 과정**

- 변환 연산 : 좌표 변환, 선형 투영, 이동, 비선형 연산 등
- 머신러닝 알고리즘은 이런 변환을 찾기 위한 창의력은 없음
- 가설 공간이라 부르는 미리 정의된 연산의 모음들을 자세히 조사하는 것뿐임
- 가능성 있는 공간을 사전에 정의하고 피드백 신호의 도움을 받아 입력 데이터에 유용한 변화는 찾는 것

## 딥러닝(Deep Learning) 

- 머신러닝의 특정한 한 분야
- **연속된 층(layer)**에서 점진적으로 의미있는 표현을 배우는 것, 데이터로부터 표현을 학습하는 새로운 방식 임

**DEEP**  
- 데이터로부터 모델을 만드는 데 얼마나 많은 층을 사용했는지가 그 모델의 깊이
- **층 기반 표현 학습** 또는 **계층적 표현 학습**라고 부르기도 함  
- 딥러닝은 기본 층을 겹겹이 쌓아 올려 구성된 **신경망**이라는 모델을 사용하여 표현 층을 학습
- 신경망이란 단어는 신경생물학의 용어
- 딥러닝의 일부 핵심 개념이 뇌구조를 이해하는 것에서부터 영감을 얻어 개발된 부분이 있으나, 뇌를 모델링한 것이 아님. 그냥 데이터로부터 표현을 학습하는 **수학 모델**임

**층(layer)**
- 신경망의 핵심 구성 요소
- 데이터 처리 필터 기능을 함
- 입력된 데이터로 부터 주어진 문제에 더 의미있는 표현(representation) 추출

- 딥러닝 모델은 데이터 정제 필터(층)가 연속되어 있는 데이터 프로세싱을 위한 여과기와 같음

**예. 숫자 분류를 위한 심층 신경망**

- 심층 신경망을 정보가 연속된 필터(filter)를 통과하면서 순도 높게 정제되는 다단계 정보 추출 작업으로 생각할 수 있음

**딥러닝은 기술적으로 데이터 표현을 학습하기 위한 다단계 처리 방식을 말함**

## 딥러닝의 작동 원리

**작동원리1. 신경망은 가중치를 파라미터로 가진다**

- 심층 신경망은 수천만 개의 파라미터를 가지기도 함

**작동원리2. 손실함수가 신경망의 출력 품질을 측정한다**

- 신경망의 출력을 제어하려면 출력이 기대하는 것보다 얼마나 벗어났는지를 측정해야 함
- 신경망의 **손실 함수(loss function) 또는 목적 함수(objective function)**가 담당
- 신경망이 한 샘플에 대해 얼마나 잘 예측했는지 측정하기 위해 손실 함수가 신경망의 예측과 진짜 타깃(신경망의 출력으로 기대하는 값)의 차이를 점수로 계산함

**작동원리3. 손실점수를 피드백 신호로 사용하여 가중치를 조정한다**

- 손실 점수를 피드백 신호로 사용하여 현재 샘플의 손실 점수가 감소되는 방향으로 가중치 값을 조금씩 수정
- 이런 수정 과정은 딥러닝의 핵심 알고리즘인 **역전파(Backpropagation) 알고리즘을 구현한 옵티마이저(optimizer)**가 담당함

- 초기에는 네트워크의 가중치가 랜덤한 값으로 할당
- 네트워크가 모든 샘플을 처리하면서 가중치가 조금씩 올바른 방향으로 조정되고 손실 점수가 감소하도록 **반복 훈련(training loop)**을 수행
- 충분한 횟수만큼 반복하면(일반적으로 수천 개의 샘플에서 수십 번 반복하면) 손실 함수를 최소화하는 가중치 값을 산출함
- 최소한의 손실을 내는 네트워크가 타깃에 가능한 가장 가까운 출력을 만드는 모델이 됨

### 딥러닝의 특징
1. 데이터의 좋은 표현을 만드는 특성 공학(feature enginerring)을 자동화함
- 특성을 직접 찾는 대신 한번에 모든 특성을 학습할 수 있음
- 머신러닝 작업 흐름을 매우 단순화함
- 고도의 다단계 작업 과정을 하나의 간단한 엔드-투-엔드(end-to-end) 딥러닝 모델로 대체할 수 있음

2. 층을 거치면서 점진적으로 더 복잡한 표현이 만들어진다는 것과 이런 점진적인 중간 표현이 공동으로 학습됨


3. 각 층은 상위 층과 하위 층의 표현이 변함에 따라서 함께 바뀜

### 왜 지금 딥러닝일까?

**딥러닝이 지속될 수 있는 특징**

# 2. 퍼셉트론과 오차역전파

### 2.1 퍼셉트론(perceptron)  


- 신경망을 이루는 가장 중요한 기본 단위
- 입력값과 활성화 함수를 사용해 출력값을 다음으로 넘기는 가장 작은 신경망 단위
- 가중치(weight)는 입력신호가 결과 출력에 주는 영향도를 조절하는 매개변수

뉴런(neuron)과 퍼셉트론(perceptron)



- 뉴런이 전기신호를 내보내 정보를 전달하는 것과 유사
- 뉴런의 수상돌기나 축색돌기처럼 신호를 전달하는 역할을 퍼셉트론에서는 **가중치(weight)**가 담당
    - 가중치가 각각의 입력신호에 부여되어 입력신호와의 계산을 하고 
    - 신호의 총합이 정해진 임계값을 넘으면 다음 뉴런으로 신호를 전달하고(뉴런의 활성화)
    - 아니면 아무것도 수행하지 않음 
        
- 각 노드의 가중치와 입력치를 곱한 것을 모두 합한 값이 활성함수에 의해 판단됨

**활성화 함수(activation function)**
- 가중합의 결과를 놓고 1 또는 0을 출력해서 다음으로 보냄
- 여기서 0과 1을 판단하는 함수

### 2.3 다층 퍼셉트론(Muti-Layer Perceptron)
- 이를 가능하게 하려면 숨어있는 층, 즉 은닉층(hidden layer)을 만들면 됨
- 은닉층이 좌표 평면을 왜곡시키는 결과를 가져옴
- 은닉층을 만들어 공간을 왜곡하면 두 영역을 가로지르는 선이 직선으로 바뀜



**은닉층의 공간 왜곡**

### 2.4 다층 퍼셉트론의 설계

- 입력층, 은닉층, 출력층



**입력층 -> 은닉층 -> 출력층**으로 결과값이 전달 됨

**XOR 문제 해결**



**numpy로 XOR 문제해결**

```python
import numpy as np

w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 =np.array([1,1])
b1 = 3
b2 = -1
b3 = -1


# 다층 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else :
        return 1

def NAND(x1, x2):
    X = np.array([x1, x2])
    return MLP(X, w11, b1)

def OR(x1, x2):
    X = np.array([x1, x2])
    return MLP(X, w12, b2)


def AND(x1, x2):
    X = np.array([x1, x2])
    return MLP(X, w2, b3)


def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))

if __name__ == '__main__':
    X = [[0,0], [0,1], [1,0], [1,1]]
    print('## XOR ##')
    for x in X:
        y = XOR(x[0], x[1])
        print('입력값',x,',출력값',y)
    print('## AND ##')
    for x in X:
        y = AND(x[0], x[1])
        print('입력값',x,',출력값',y)
    print('## OR ##')
    for x in X:
        y = OR(x[0], x[1])
        print('입력값',x,',출력값',y)
```

### 2.5 오차 역전파(Backpropagation)

**단일 퍼셉트론의 가중치 조정**

**다층 퍼셉트론의 가중치 조정**

**최적화의 계산 방향이 출력층에서 앞의 층으로 진행 = 오차 역전파(back propagation)**

**오차 역전파 동작 단계**
1. 임의의 초기 가중치(w(1))를 준 뒤 결과(y_out)를 계산 
2. 계산 결과(예측값)와 원하는 값(실제값,타깃값) 사이의 오차 계산 
3. 경사 하강법을 이용해 바로 앞 가중치를 **오차가 작아지는 방향으로 업데이트**

**오차가 작아지는 방향으로 업데이트 한다**

- 미분 값이 0에 가까워지는 방향으로 나간다
- 기울기가 0이 되는 방향


### 오차역전파 관련 파이썬 코드


- 신경망 클래스 구현

```python
#### 오차역전파를 이용한 신경망 구현 : XOR 문제

import random
import numpy as np

##### 신경망 클래스 구현 #############################

random.seed(777)

class NeuralNetwork:

    # 초깃값의 지정
    def __init__(self, num_x, num_yh, num_yo, bias=1):

        # 입력값(num_x), 은닉층 초깃값(num_yh), 출력층 초깃값(num_yo), 바이어스
        self.num_x = num_x + bias  # 바이어스는 1로 지정(본문 참조)
        self.num_yh = num_yh
        self.num_yo = num_yo

        # 활성화 함수 초깃값
        self.activation_input = [1.0] * self.num_x
        self.activation_hidden = [1.0] * self.num_yh
        self.activation_out = [1.0] * self.num_yo

        # 가중치 입력 초깃값
        self.weight_in = makeMatrix(self.num_x, self.num_yh)
        for i in range(self.num_x):
            for j in range(self.num_yh):
                self.weight_in[i][j] = random.random()

        # 가중치 출력 초깃값
        self.weight_out = makeMatrix(self.num_yh, self.num_yo)
        for j in range(self.num_yh):
            for k in range(self.num_yo):
                self.weight_out[j][k] = random.random()

        # 모멘텀 SGD를 위한 이전 가중치 초깃값
        self.gradient_in = makeMatrix(self.num_x, self.num_yh)
        self.gradient_out = makeMatrix(self.num_yh, self.num_yo)

    # 업데이트 함수
    def update(self, inputs):

        # 입력 레이어의 활성화 함수
        for i in range(self.num_x - 1):
            self.activation_input[i] = inputs[i]

        # 은닉층의 활성화 함수
        for j in range(self.num_yh):
            sum = 0.0
            for i in range(self.num_x):
                sum = sum + self.activation_input[i] * self.weight_in[i][j]
            # 시그모이드와 tanh 중에서 활성화 함수 선택
            self.activation_hidden[j] = tanh(sum, False)

        # 출력층의 활성화 함수
        for k in range(self.num_yo):
            sum = 0.0
            for j in range(self.num_yh):
                sum = sum + self.activation_hidden[j] * self.weight_out[j][k]
            # 시그모이드와 tanh 중에서 활성화 함수 선택
            self.activation_out[k] = tanh(sum, False)

        return self.activation_out[:]
    
    # 역전파의 실행
    def backPropagate(self, targets):

        # 델타 출력 계산
        output_deltas = [0.0] * self.num_yo
        for k in range(self.num_yo):
            error = targets[k] - self.activation_out[k]
            # 시그모이드와 tanh 중에서 활성화 함수 선택, 미분 적용
            output_deltas[k] = tanh(self.activation_out[k], True) * error

        # 은닉 노드의 오차 함수
        hidden_deltas = [0.0] * self.num_yh
        for j in range(self.num_yh):
            error = 0.0
            for k in range(self.num_yo):
                error = error + output_deltas[k] * self.weight_out[j][k]
                # 시그모이드와 tanh 중에서 활성화 함수 선택, 미분 적용
            hidden_deltas[j] = tanh(self.activation_hidden[j], True) * error

        # 출력 가중치 업데이트
        for j in range(self.num_yh):
            for k in range(self.num_yo):
                gradient = output_deltas[k] * self.activation_hidden[j]
                v = mo * self.gradient_out[j][k] - lr * gradient
                self.weight_out[j][k] += v
                self.gradient_out[j][k] = gradient

        # 입력 가중치 업데이트
        for i in range(self.num_x):
            for j in range(self.num_yh):
                gradient = hidden_deltas[j] * self.activation_input[i]
                v = mo*self.gradient_in[i][j] - lr * gradient
                self.weight_in[i][j] += v
                self.gradient_in[i][j] = gradient

        # 오차의 계산(최소 제곱법)
        error = 0.0
        for k in range(len(targets)):
            error = error + 0.5 * (targets[k] - self.activation_out[k]) ** 2
        return error

    # 학습 실행
    def train(self, patterns):
        for i in range(iterations):
            error = 0.0
            for p in patterns:
                inputs = p[0]
                targets = p[1]
                self.update(inputs)
                error = error + self.backPropagate(targets)
            if i % 500 == 0:
                print('error: %-.5f' % error)
                
    # 결괏값 출력
    def result(self, patterns):
        for p in patterns:
            print('Input: %s, Predict: %s' % (p[0], self.update(p[0])))
```

- 활성화함수 정의

```python
# 활성화 함수 - 1. 시그모이드
# 미분할 때와 아닐 때의 각각의 값
def sigmoid(x, derivative=False):
    if (derivative == True):
        return x * (1 - x)
    return 1 / (1 + np.exp(-x))

# 활성화 함수 - 2. tanh
# tanh 함수의 미분은 1 - (활성화 함수 출력의 제곱)
def tanh(x, derivative=False):
    if (derivative == True):
        return 1 - x ** 2
    return np.tanh(x)

# 가중치 배열 만드는 함수
def makeMatrix(i, j, fill=0.0):
    mat = []
    for i in range(i):
        mat.append([fill] * j)
    return mat
```

### 2.6. 기울기 소실 문제와 활성화 함수

#### 다층 확장 오차 역전파
- 출력층으로부터 하나씩 앞으로 되돌아가며 각 층의 가중치를 수정
- 층이 늘어나면서 기울기가 중간에 0이 되어버리는 **기울기 소실(vanishing gradient)** 문제가 발생

- 활성화 함수로 사용된 시그모이드(sigmoid) 함수의 특성으로  
- 여러 층을 거칠수록 기울기가 사라져 가중치를 수정하기가 어려워지는 것

**이를 해결하고자 활성화 함수를 시그모이드가 아닌 여러 함수로 대체하기 시작함**

**하이퍼볼릭 탄젠트(tanh) 함수**

- 시그모이드 함수의 범위를 -1에서 1로 확장
- 미분시 1보다 작은 값이 존재하여 기울기 소실 문제는 사라지지 않음

**렐루(ReLU) 함수**
- 0미만의 값은 모두 0으로 처리하고, 0 이상의 값은 원본 값 그대로 출력  
- 현재 가장 많이 사용되는 활성화 함수
- 토론토대학교의 제프리 힌튼 교수
- 여러 은닉층을 거치며 곱해지더라도 맨 처음 층까지 사라지지 않고 남아있을 수 있음  
- 여러 층을 쌓을 수 있게 함으로써 딥러닝의 발전에 속도가 붙게 됨 

**소프트플러스(softplus) 함수**
- 렐루의 0이 되는 순간을 완화

### 2.7 경사하강법과 옵티마이저

- 속도와 정확도 문제를 해결

**케라스의 optimizer 파라미터 인수와 공식**

1.확률적 경사하강법(SGD)
- 속도 개선
- `keras.optimizers.SGD(lr=0.1)`

2.모멘텀(momentum)
- 관성, 탄력, 가속도
- 경사 하강법에 탄력을 주는 것
- 오차를 수정하기 전 바로 앞 수정 값과 방향(+, -)을 참고하여 같은 방향으로 일정한 비율만 수정하는 방법
- 이전 이동 값을 고려하여 일정 비율만큼만 다음 값을 결정하므로 관성의 효과를 낼 수 있음
-` keras.optimizers.SGD(lr=0.1, momentum=0.9)`

3.네스테로프 모멘텀(Nesterov Accelrated Gradient:NAG)
- 모멘텀이 이동시킬 방향으로 미리 이동해서 그레디언트를 계산
- 불필요한 이동을 줄이는 효과
- `keras.optimizers.SGD(lr=0.1, momentum=0.9, nesterov=True)`

4.아다그라드(Adagrad: Adaptive Gradient)
- 변수의 업데이트가 잦으면 학습률을 적게하여 이동 보폭을 조절하는 방법
- `keras.optimizers.Adagrad(lr=0.01, epsilon=1e-6)`

5.알엠에스프롭(RMSprop)
- 아다그라드의 보폭 민감도를 보완하는 방법
- `keras.optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)`

6.아담(Adam:Adaptive Moment Estimation)
- 모멘텀과 알엠에스프롭 방법을 결합한 방법
- 자주 사용
- `keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.9999, epsilon=1e-08, decay=0.0)`
