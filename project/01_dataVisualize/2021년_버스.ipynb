```python
import pandas as pd
```

```python
list = ['0'+str(i) for i in range(1,10)] + ['10', '11', '12']

df_A = pd.read_csv(f"raw_data/2021년_버스노선별_정류장별_시간대별_승하차_인원_정보(01월).csv", encoding='cp949')
for i in list:
    month = i
    df_B = pd.read_csv(f"raw_data/2021년_버스노선별_정류장별_시간대별_승하차_인원_정보({month}월).csv", encoding='cp949')
    df_A = pd.concat([df_A,df_B])

df_A
```

```python
df = df_A[["사용년월", "노선번호", "노선명", "표준버스정류장ID", '역명', '7시하차총승객수', '8시하차총승객수', '9시하차총승객수', '10시하차총승객수', '17시승차총승객수', '18시승차총승객수', '19시승차총승객수', '20시승차총승객수']]
```

```python
df
```

```python
pd_pivot = pd.pivot_table(df,index=['사용년월','노선번호','노선명','표준버스정류장ID','역명'], aggfunc='sum')
pd_pivot
```

```python
pd_pivot.reset_index(inplace=True)
```
